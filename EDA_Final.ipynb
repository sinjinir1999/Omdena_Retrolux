```python
import numpy as np
mean = (df[df['ROI']!=np.inf]).ROI.mean()
max_ = (df[df['ROI']!=np.inf]).ROI.max()
min_ = (df[df['ROI']!=np.inf]).ROI.min()
mean, max_, min_

print('Min of ROI is : ',min_ )
print('Mean of ROI is : ',round(mean,2) )
print('Max of ROI is : ', round(max_,2) )
```

```python
no_inf = df[(df['ROI']!=np.inf ) ]
no_inf.shape
```

```python
no_nan_inf = no_inf[no_inf.ROI.notnull()]
no_nan_inf.shape
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots()
ax.ticklabel_format(useOffset=False, style='plain')
ax.plot(no_nan_inf['ROI'].value_counts())

```

```python
q1 = np.quantile(no_nan_inf.ROI, 0.25)
 
q3 = np.quantile(no_nan_inf.ROI, 0.75)
med = np.median(no_nan_inf.ROI)
 
iqr = q3-q1
 
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(iqr, upper_bound, lower_bound)
```

```python
outliers = no_nan_inf.ROI[(no_nan_inf.ROI <= lower_bound) | (no_nan_inf.ROI >= upper_bound)]
print('The Number of outliers in the boxplot: ',len(outliers))
print()
print('The following are the outliers in the boxplot:{}'.format(outliers))
```

```python
outliers.head(50)
```

```python
arr2 = no_nan_inf.ROI[(no_nan_inf.ROI >= lower_bound) & (no_nan_inf.ROI <= upper_bound)]
plt.figure(figsize=(12, 7))
plt.boxplot(arr2)
plt.show()
```

```python
x=no_nan_inf[(no_nan_inf.ROI >= lower_bound) & (no_nan_inf.ROI <= upper_bound)]
mean = (x[x['ROI']!=np.inf]).ROI.mean()
max_ = (x[x['ROI']!=np.inf]).ROI.max()
min_ = (x[x['ROI']!=np.inf]).ROI.min()
mean, max_, min_

print('Min of ROI is : ',min_ )
print('Mean of ROI is : ',round(mean,2) )
print('Max of ROI is : ', round(max_,2) )
```
