# Description of extracted features

- borough code: Bronx(BX), Brooklyn(BK), Manhattan(MN), Queens(QN), Staten Island = SI
- zipcode: zip code
- address:  The general format is house number or low house number, if there is a house number range, and a space followed by a street name
- bldgclass(Building Class): A code describing the major use of structures
- landuse(land use category): A code for the land use category
- ownertype: A code indicating type of ownership
- ownername: The name of the owner of the lot
- lotarea: Total area of the lot, expressed in square feet rounded to the nearest integer.
- bldgarea(floor area, total building): The total gross area in square feet.
- comarea(floor area, commercial): An estimate of the exterior dimensions of the portion of the structure(s) allocated for commercial use
- officearea(floor area, office): An estimate of the exterior dimensions of the portion of the structure(s) allocated for office use
- retailarea(floor area, retail): An estimate of the exterior dimensions of the portion of the structure(s) allocated for real use
- garagearea(floor area, garage): An estimate of the exterior dimensions of the portion of the structure(s) allocated for garage use.
- strgearea(floor area, storage): An estimate of the exterior dimensions of the portion of the structure(s) allocated for storage or loft purposes
- numbldgs(number of buildings): The number of buildings on the lot.
- numfloors(number of floors): In the tallest building on the lot, the number of full and partial stories starting from the ground floor.
- unitres(residential units) : The sum of residential units in all buildings on the tax lot.
- unitstotal( resisdential and non-residential units): The sum of residential and non-residential (offices, retail stores, etc.) units in all buildings on the lot.
- yearbuilt: The year construction of the building was completed.
- commfar: maximum allowable commercial floor area ratio
- xcoord: The X coordinate of the XY coordinate pair which depicts the approximate location of the lot.
- ycoord: The Y coordinate of the XY coordinate pair which depicts the approximate location of the lot.
