```python
#x=df.isnull()
x = df[df.notnull()]
x.shape
```

```python
corr = x.corr().abs().unstack().sort_values(kind="quicksort", ascending=False).reset_index()
corr.rename(columns={"level_0": "Feature 1", "level_1": "Feature 2", 0: 'Correlation Coefficient'}, inplace=True)
corr[corr['Feature 1'] == 'payback']
```

```python
from sklearn.model_selection import train_test_split

columns = ['ROI','project_id','option_id','option_calculation_id','company_id','sol_total_existing_count','total_product_count','energy_savings_per_year']
X = x[columns]
y = x['payback']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)

pred = linreg.predict(X_test)


print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(linreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(linreg.score(X_test, y_test)))
```
