
### 4. Data Pre-processing for Existing Lighting and Existing Fixtures 

---






```python
# General Description of the DataFrame
df_existing_lighting.head() #view the first 5 rows of the dataframe
```

```python
# General Description of the DataFrame
df_existing_fixtures.head() #view the first 5 rows of the dataframe
```

Table contains 41 columns and 77546 rows.

```python
# Get information about the Existing Fixtures Table
df_existing_fixtures.info()
```

Table contains 99 columns and 190876 rows


```python
# Drop irrelevant columns and keep important columns. Rename id in existing lighting table as 'existing_lighting_id' foreign key
df_existing_lighting_new = df_existing_lighting[['id','name','description','watts_per_product','lm70','project_id','product_type_id','technology_id','category_id','subcategory_id','active','inactive_at','power_factor','existing_product_type']]
df_existing_fixtures_new = df_existing_fixtures[['id','area_id','operating_schedule_id','existing_count','existing_lighting_id','active','year_installed','working_fixtures_count']]
df_existing_lighting_new.rename(columns= {'id':'existing_lighting_id'}, inplace = True) # rename id to existing_lighting_id
df_existing_lighting_new.info()
```

Table columns has been reduced from 41 to 13.


```python
df_existing_fixtures_new.info()
```

Table has been reduced from 99 columns to 8 columns

```python
#Ensure that it is only active fixtures and lighting that is contained in the table i.e active = True (t)
df_existing_lighting_new = df_existing_lighting_new[df_existing_lighting_new.active== 't']
df_existing_fixtures_new = df_existing_fixtures_new[df_existing_fixtures_new.active == 't']
```

```python
df_existing_lighting_new.info()
```

Roughly 3% of the existing lighting were inactive

```python
df_existing_fixtures_new.info()
```

Approximately 8% of the existing fixtures were inactive.

```python
#Check number of unique entries in the Existing Lighting Table 
df_existing_lighting_new.nunique()
```

```python
# Merge the existing lighting table to the existing fixtures table using the existing_lighting_id as a foreign key.
df_existing_fixtures_lighting = pd.merge(df_existing_fixtures_new,df_existing_lighting_new, how = 'left',on = 'existing_lighting_id' )
df_existing_fixtures_lighting.head()
```

```python
#Check number of columns in the new table  
df_existing_fixtures_lighting.info()
```

We may need to drop the Product_type, technology, category and subcategory foreign keys as they do not contain any data in this new table.

```python
# Drop most of the empty columns
df_existing_fixtures_lighting = df_existing_fixtures_lighting.drop(columns = ['product_type_id','technology_id','category_id','subcategory_id','inactive_at','power_factor'])
df_existing_fixtures_lighting.info()
```

### 6. Data Pre-processing for Existing_Fixtures_Lighting_Schedule and Rate_Schedule


```python
# Description of the Rate Schedule Table
df_rate_schedule.info()
```

```python
df_rate_schedule.head()
```

```python
# Ensure it is only active projects we are using and drop all the irrelevant columns
df_rate_schedule_new =df_rate_schedule[df_rate_schedule.active =='t']
df_rate_schedule_new = df_rate_schedule_new.drop(columns = ['inactive_at','created_by_user_id'])
df_rate_schedule_new.info() 
```

```python
# Rename the id to operating_schedule_id 
df_rate_schedule_new.rename(columns = {'id':'rate_schedule_id'}, inplace = True)
df_rate_schedule_new.info()
```

```python
# Drop Project_id with least data points in Existing_fistures_lighting_schedule to avoid duplication 
df_existing_fixtures_lighting_schedule.drop(columns=['project_id_y'], inplace = True)
df_existing_fixtures_lighting_schedule.rename(columns={'project_id_x':'project_id'},inplace = True)
df_existing_fixtures_lighting_schedule.info()
```

```python
#Check number of unique entries in rate_schedule_new table
df_rate_schedule_new.nunique()
```

The table contains duplicate 'project_ids'. So we keep first occurrence and remove other duplicates inorder to have a fact table for merging with other table

```python
df_rate_schedule_new.drop_duplicates(['project_id'],keep = 'first',inplace = True)
df_rate_schedule_new.info()
```

```python
df_existing_fixtures_lighting_schedule.nunique()
```

```python
df_existing_fixtures_lighting_schedule = pd.merge(df_existing_fixtures_lighting_schedule,df_rate_schedule_new, how ='left' , on = 'project_id')
df_existing_fixtures_lighting_schedule.info()
```

```python
# Remove all the irrelevant columns and rename some of the columns for easy identification
df_existing_fixtures_lighting_schedule.drop(columns=['active_y','created_at_x','created_at_y','updated_at_x','updated_at_y','inactive_at','copied_from_id_x','copied_from_id_y','uuid_x','uuid_y'],inplace=True)
df_existing_fixtures_lighting_schedule.rename(columns={'name_x':'name_lighting','name_y':'name_operating','name':'name_rate','active_x':'active_fixtures','notes_x':'notes_operating','notes_y':'notes_rate','internal_notes_x':'internal_notes_operating','internal_notes_y':'internal_notes_rate'}, inplace = True)
df_existing_fixtures_lighting_schedule.info()
```
